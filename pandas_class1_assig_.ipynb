
# Q7. DataFrame from NumPy Array

## Problem Statement
Create the following NumPy array:

```python
[[1,2],
 [3,4],
 [5,6]]
```

Convert it into a Pandas DataFrame.

## Sample Output
```python
   0  1
0  1  2
1  3  4
2  5  6
```


````python
import numpy as np
import pandas as pd
arr = np.array([
     [1,2],
     [3,4],
     [5,6]
    ])

df = pd.DataFrame(arr)

print(df)
````
